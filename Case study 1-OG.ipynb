## Data Cleaning

```python
# Reading CSV file
df_bank = pd.read_csv(
    './data/bank-additional.csv',
    delimiter=';',
    na_values=['unknown',]
)

# Convert pdays to a categorical column
df_bank['pdays_cat'] = np.select(
    [
        df_bank['pdays'] == 999, 
        (df_bank['pdays'] >= 1) & (df_bank['pdays'] <= 7),
        (df_bank['pdays'] >= 8) & (df_bank['pdays'] <= 14),
        (df_bank['pdays'] >= 15) & (df_bank['pdays'] <= 21)
    ], 
    [
        'NEVER CALLED OR ONLY CALLED ONCE', 
        '0-1 WEEKS BETWEEN LAST 2 CALLS',
        '1-2 WEEKS BETWEEN LAST 2 CALLS',
        '2-3 WEEK BETWEEN LAST 2 CALLS'
    ], 
    default = np.nan
)

#recode y to be binary 
### 1 = YES, 0 = NO 
df_bank['y'] = np.where(df_bank['y'].str.lower() == 'yes', 1, 0)


# drop these columns
df_bank = df_bank.drop(
    ['default', 'duration', 'pdays', 'month','day_of_week'],
    axis = 1
)

# define non-numeric columns
categorical_cols = ["job", "marital", "education", "housing", "loan", "contact",
                    "poutcome",  'pdays_cat']

# convert categorical columns to dummies, kill one of them w/ drop = true
df_bank = pd.get_dummies(df_bank, columns=categorical_cols,drop_first=True)
```

## Dealing with Imbalance by Trimming the data

#### Lets reset the data

```python
# Reading CSV file
df_bank = pd.read_csv(
    './data/bank-additional.csv',
    delimiter=';',
    na_values=['unknown',]
)

# Creating pdays_cat column
df_bank['pdays_cat'] = np.select(
    [
        df_bank['pdays'] == 999, 
        (df_bank['pdays'] >= 1) & (df_bank['pdays'] <= 7),
        (df_bank['pdays'] >= 8) & (df_bank['pdays'] <= 14),
        (df_bank['pdays'] >= 15) & (df_bank['pdays'] <= 21)
    ], 
    [
        'NEVER CALLED OR ONLY CALLED ONCE', 
        '0-1 WEEKS BETWEEN LAST 2 CALLS',
        '1-2 WEEKS BETWEEN LAST 2 CALLS',
        '2-3 WEEK BETWEEN LAST 2 CALLS'
    ], 
    default = np.nan
)

#recode y to be binary 
### 1 = YES, 0 = NO 
df_bank['y'] = np.where(df_bank['y'].str.lower() == 'yes', 1, 0)


# columns to drop
df_bank = df_bank.drop(
    ['default', 'duration', 'pdays', 'month','day_of_week'],
    axis = 1
)

categorical_cols = ["job", "marital", "education", "housing", "loan", "contact",
                    "poutcome",  'pdays_cat']

df_bank = pd.get_dummies(df_bank, columns=categorical_cols,drop_first=True)
```

```python
positive_samples = df_bank[df_bank['y'] == 1]  # Select all rows where 'y' is 1
negative_samples = df_bank[df_bank['y'] == 0]  # Select all rows where 'y' is 0

# Randomly sample 451 rows where 'y' is 0
random_negative_samples = negative_samples.sample(n=451, random_state=42)

# Combine the positive and negative samples
final_sample = pd.concat([positive_samples, random_negative_samples])
```

```python
X = final_sample.drop('y', axis=1)
y = final_sample['y']
calculate_vif(X, y)
```
